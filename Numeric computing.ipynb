```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (10,8)
```

```python
# Euler methot or can say numeric method for solve differantial equation

list_vx = []
list_x = []
list_vy = []
list_y = []
list_t = []


T = 10
m = 1
L = 5
g = 9.81
beta = 0.05
gamma = T/(L*m)


h = 0.1
t_0 = 0
v_0x = -1
v_0y = -1
x_0 = 3
y_0 = -4


n = int(100/h)

for i in range(n):

    t_i = h + t_0  # time variables

    '''
    Computing for equation with x component
    '''
    df1x = v_0x # there is not any variable which dependence x or t variables
    df2x = -gamma * x_0 # we have one dependence variables.

    x_i = x_0 + h * df1x # take a step for x variables
    v_ix = v_0x + h * df2x # take a step for v variables

    v_0x = v_ix
    x_0 = x_i


    '''
    Computing for equation with y component
    '''
    df1y =  v_0y  # there is not any variable which dependence x or t variables
    df2y = - gamma * y_0 - g - beta * np.sin(2*np.pi*t_i)# we have one dependence variables.

    y_i = y_0 + h * df1y  # take a step for x variables
    v_iy = v_0y + h * df2y  # take a step for v variables

    v_0y = v_iy
    y_0 = y_i

    t_0 = t_i


    list_vx.append(round(v_0x, 2))
    list_x.append(round(x_0,2))
    list_vy.append(round(v_0y, 2))
    list_y.append(round(y_0, 2))
    list_t.append(round(t_0,2))



print(f"length v_x: {len(list_vx)}")
print(f"length x: {len(list_x)}")
print(f"length v_y: {len(list_vy)}")
print(f"length y: {len(list_y)}")
print(f"length t: {len(list_t)}")

#print(f"list v_x: {list_vx}")
#print(f"list x: {list_x}")
#print(f"list v_y: {list_vy}")
#print(f"list y: {list_y}")
#print(f"list t: {list_t}")
```

```python
plt.subplot(3,2,1)
plt.plot(list_vx, list_vy)
plt.title("Velocity phase space")
plt.xlabel("$v_x$")
plt.ylabel("$v_y$")

plt.subplot(3,2,2)
plt.plot(list_x, list_y)
plt.title("Coordinates phase space")
plt.xlabel("$x$")
plt.ylabel("$y$")

plt.subplot(3,2,3)
plt.plot(list_t, list_x)
plt.title("Curve x coordinates relative to time")
plt.xlabel("$t$")
plt.ylabel("$x$")

plt.subplot(3,2,4)
plt.plot(list_t, list_y)
plt.title('Curve y coordinates relative to time')
plt.xlabel("$t$")
plt.ylabel("$y$")

plt.subplot(3,2,5)
plt.plot(list_x, list_vx)
plt.title("$v_x$ to $x$ phase space")
plt.xlabel("$x$")
plt.ylabel("$v_x$")

plt.subplot(3,2,6)
plt.plot(list_y, list_vy)
plt.title("$v_y$ to $y$ phase space")
plt.xlabel("$y$")
plt.ylabel("$v_y$")


plt.tight_layout()
plt.show()
```

```python
dict_d = {
    't': list_t,
    'x': list_x,
    'y': list_y,
    'v_x': list_vx,
    'v_y': list_vy
}
```
