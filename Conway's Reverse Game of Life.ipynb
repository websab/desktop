## Look at the img

```python
def select_pic(i, row, col):
    arr = np.array(train.iloc[i:i+1,2:402])
    pix = arr.reshape(-1, 20, 20)
    img = pix[0]
    plt.subplot(row, col, i+1)
    plt.imshow(img, cmap='gray')
```

```python
plt.figure(figsize=(10,10))
for i in range(100):
    select_pic(i,10,10)
plt.show()
```
